```python
import os, json

BASEDIR = "output/degree_e2e_ace05ep/multi-task-Keywords-full-new"
ee_predictions = None
eke_predictions = None

with open(os.path.join(BASEDIR, "eval_result.json"), "r") as f:
    ee_predictions = json.load(f)

with open(os.path.join(BASEDIR, "keywords_result.json"), "r") as f:
    eke_predictions = json.load(f)
```

```python
contain_events_ids = []

for idx in range(len(ee_predictions)):
    if ee_predictions[idx]['pred triggers'] or ee_predictions[idx]['gold triggers'] or ee_predictions[idx]['pred roles'] or ee_predictions[idx]['gold roles']:
        contain_events_ids.append(idx)

len(contain_events_ids)
```

```python
def get_ee_record(idx):
    return ee_predictions[idx]

def get_eke_record(idx):
    return eke_predictions[idx]
```

```python
# 1. , 
correct_ee_ids = []
for idx in contain_events_ids:
    if (len(ee_predictions[idx]['pred triggers']) == len(ee_predictions[idx]['gold triggers']) and
        len(ee_predictions[idx]['pred roles']) == len(ee_predictions[idx]['gold roles']) and len(ee_predictions[idx]['input text'].split()) < 20):
        correct_ee_ids.append(idx)

len(correct_ee_ids)
```

```python
# final example
get_ee_record(correct_ee_ids[21])
```

```python
# 1. , 
part_correct_ee_ids = []
for idx in contain_events_ids:
    if len(ee_predictions[idx]['input text'].split()) < 20:
        if len(ee_predictions[idx]['pred triggers']) != len(ee_predictions[idx]['gold triggers']):
            part_correct_ee_ids.append(idx)
        # if ((len(ee_predictions[idx]['pred triggers']) != len(ee_predictions[idx]['gold triggers']) or
        #     len(ee_predictions[idx]['pred roles']) != len(ee_predictions[idx]['gold roles'])) and len(ee_predictions[idx]['input text'].split()) < 30):
        #     correct_ee_ids.append(idx)

len(part_correct_ee_ids)
```

```python
#  vote  Elect 
get_ee_record(part_correct_ee_ids[0])
```

```python
# 
extra_keywords_eke_ids = []
for idx in contain_events_ids:
    eke_record = get_eke_record(idx)
    if len(eke_record['input text'].split()) > 20:
        g_texts = eke_record['gold text']
        p_texts = eke_record['pred text']
        for pt, gt in zip(p_texts, g_texts):
            if "<Keyword>" in pt and "<Keyword>" not in gt:
                extra_keywords_eke_ids.append(idx)
                break

len(extra_keywords_eke_ids)
```

```python
get_eke_record(extra_keywords_eke_ids[1])
```

```python
# 
def get_label_num(text):
    cnt = 0
    start = 0
    while True:
        start = text.find("<Keyword>", start)
        if start == -1:
            break
        start += len("<Keyword>")
        cnt += 1

    return cnt

miss_keywords_eke_ids = []
for idx in contain_events_ids:
    eke_record = get_eke_record(idx)
    if len(eke_record['input text'].split()) < 20:
        g_texts = eke_record['gold text']
        p_texts = eke_record['pred text']
        for pt, gt in zip(p_texts, g_texts):
            if get_label_num(gt) > get_label_num(pt):
                miss_keywords_eke_ids.append(idx)
                break

len(miss_keywords_eke_ids)
```
